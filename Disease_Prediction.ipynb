```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## EDA

```python
df = pd.read_csv('./DiseaseAndSymptoms.csv')
df.head()
```

```python
Y = df['Disease']
```

```python
Y
```

```python
df.drop('Disease',axis=1,inplace=True)
```

```python
symptoms = set()
for col in df.columns:
    for symptom in df[col].unique():
        symptoms.add(symptom)
symptoms
```

```python
symptoms.remove(np.nan)
symptoms
```

```python
len(symptoms)
```

```python
data = np.zeros((4920,131)).astype(int)
data
```

```python
df_final = pd.DataFrame(data,columns=list(symptoms))
df_final['Disease'] = Y
df_final
```

```python
df['Symptom_1'][2]
```

```python
for i in range(len(df)):
    data = df.iloc[i].values
    for symptom in data:
        df_final.loc[i,symptom] = 1
df_final
```

```python
df_final.drop(df_final.columns[-1],axis=1,inplace=True)
df_final
```

## Train Model

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df_final[df_final.columns[:-1]],df_final[df_final.columns[-1]],test_size=0.3)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
```

### MNB

```python
mnb = MultinomialNB()
mnb.fit(x_train,y_train)

print("Score: ",mnb.score(x_test,y_test))

cm = confusion_matrix(mnb.predict(x_test),y_test)
sns.heatmap(cm,annot=True)
```

## Random Forest

```python
rf = RandomForestClassifier(oob_score=True, bootstrap=True)
rf.fit(x_train,y_train)

print("Score: ",rf.score(x_test,y_test))

cm = confusion_matrix(rf.predict(x_test),y_test)
sns.heatmap(cm,annot=True)
```
